```python
import pandas as pd
import networkx as nx
```

```python
reactomeX = nx.read_gpickle('../output/reactomeX_v1.1.gpickle')
keggX = nx.read_gpickle('../output/KEGGeractome_v1.1.gpickle')
```

```python
print(nx.info(reactomeX))
```

```python
print(nx.info(keggX))
```

```python
reactomeX.remove_edges_from([(source, target) for source,target in keggX.edges()])
reactomeX.remove_edges_from([(target, source) for source,target in keggX.edges()])
```

```python
reactomX_edgelist = nx.to_pandas_edgelist(reactomeX)
reactomX_edgelist.head()
```

```python
keggX_edgelist = nx.to_pandas_edgelist(keggX)
keggX_edgelist.head()
```

```python
pathwayX_edgelist = pd.concat([reactomX_edgelist, keggX_edgelist])
pathwayX_edgelist[['effect', 'indirect']] = pathwayX_edgelist[['effect', 'indirect']].astype(int)
pathwayX_edgelist.head()
```

```python
pathwayX = nx.from_pandas_edgelist(pathwayX_edgelist, edge_attr=True, create_using=nx.DiGraph())
```
